# Long Short-Term Memory (LSTM)

Long Short-Term Memory (LSTM) is a type of recurrent neural network (RNN) architecture that was designed to model temporal sequences and their long-range dependencies more accurately than conventional RNNs.

## How it Works

LSTM works by using a series of "gates" that can control the flow of information in and out of the memory state of the cell. These gates can learn which data in a sequence is important to keep or throw away, thereby allowing it to pass relevant information down the sequence chain.

The LSTM cell contains three gates: the input gate, which adds information to the cell state; the forget gate, which removes information that is no longer required; and the output gate, which outputs the information based on the current input and the cell state.

## Use Cases

LSTMs have a wide range of applications, including:

- **Natural Language Processing (NLP)**: LSTMs are used for various NLP tasks, such as language modeling, machine translation, and sentiment analysis.

- **Speech Recognition**: LSTMs can model the temporal dependencies in speech signals, making them useful for speech recognition tasks.

- **Time Series Prediction**: LSTMs can capture long-term dependencies in time series data, making them useful for predicting future values in a time series, such as stock prices or weather patterns.

## Limitations

Despite their advantages, LSTMs also have some limitations:

- **Computational Intensity**: LSTMs can be computationally intensive to train, particularly for long sequences.

- **Need for Large Amounts of Data**: To perform well, LSTMs typically require large amounts of labeled data.

- **Interpretability**: Like other deep learning models, LSTMs can be seen as "black boxes" and can be difficult to interpret. This can make it challenging to understand why a particular prediction was made.

# Долгая краткосрочная память (LSTM)

Долгая краткосрочная память (LSTM) - это тип архитектуры рекуррентной нейронной сети (RNN), которая была разработана для моделирования временных последовательностей и их долгосрочных зависимостей более точно, чем обычные RNN.

## Как это работает

LSTM работает с использованием ряда "вентилей", которые могут контролировать поток информации внутрь и из состояния памяти ячейки. Эти вентили могут учиться, какие данные в последовательности важны для сохранения или удаления, тем самым позволяя передавать релевантную информацию по цепочке последовательностей.

Ячейка LSTM содержит три вентиля: входной вентиль, который добавляет информацию в состояние ячейки; вентиль забывания, который удаляет информацию, которая больше не требуется; и выходной вентиль, который выводит информацию на основе текущего входа и состояния ячейки.

## Варианты использования

У LSTM есть широкий спектр применений, включая:

- **Обработка естественного языка (NLP)**: LSTM используются для различных задач NLP, таких как моделирование языка, машинный перевод и анализ тональности.

- **Распознавание речи**: LSTM могут моделировать временные зависимости в речевых сигналах, что делает их полезными для задач распознавания речи.

- **Прогнозирование временных рядов**: LSTM могут захватывать долгосрочные зависимости в данных временных рядов, что делает их полезными для прогнозирования будущих значений во временном ряду, таких как цены на акции или погодные условия.

## Ограничения

Несмотря на их преимущества, у LSTM также есть некоторые ограничения:

- **Вычислительная интенсивность**: LSTM могут быть вычислительно интенсивными для обучения, особенно для длинных последовательностей.

- **Необходимость большого количества данных**: Для хорошей работы LSTM обычно требуются большие объемы размеченных данных.

- **Интерпретируемость**: Как и другие модели глубокого обучения, LSTM могут рассматриваться как "черные ящики" и могут быть сложными для интерпретации. Это может затруднить понимание причин, по которым был сделан определенный прогноз.

# Gates in LSTM

In Long Short-Term Memory (LSTM) units, three types of gates are present: the input gate, the forget gate, and the output gate. Each gate in an LSTM unit has the same size as the hidden state of the LSTM.

<li>Input Gate: This gate controls the extent of new input to be stored in the cell state. It employs a sigmoid activation function to assign a value between 0 and 1 to each piece of information in the input. A value close to 1 signifies that the information is crucial and should be stored, while a value close to 0 indicates that the information is not important and can be disregarded.

<li>Forget Gate: This gate decides the amount of information to be retained from the cell state. It also uses a sigmoid activation function to assign a value between 0 and 1 to each piece of information in the cell state. A value close to 1 indicates that the information should be retained, while a value close to 0 suggests that the information should be discarded.

<li>Output Gate: This gate controls the amount of the cell state to be output at the current time step. It uses a sigmoid activation function to assign a value between 0 and 1 to each piece of information in the cell state. A value close to 1 indicates that the information should be output, while a value close to 0 means that the information should be kept inside the cell.

These gates enable the LSTM to selectively remember or forget information, making it effective for tasks involving sequences or time series data. The gates are implemented internally by the LSTM class in PyTorch, so they are not explicitly defined in the code.

The size of the gates in an LSTM (Long Short-Term Memory) unit is determined by the size of the hidden state. Each gate in an LSTM unit has the same size as the hidden state. This means that the output of the input gate, forget gate, and output gate are all of the same size as the hidden state.

# Вентили в LSTM

В блоках долгосрочной памяти с коротким сроком действия (Long Short-Term Memory, LSTM) присутствуют три типа вентилей: входной вентиль, вентиль забывания и выходной вентиль. Каждый вентиль в блоке LSTM имеет такой же размер, как скрытое состояние LSTM.

<li>Входной вентиль: Этот вентиль контролирует, какая часть нового входа должна быть сохранена в состоянии ячейки. Он использует сигмоидную функцию активации для присвоения значения от 0 до 1 каждому элементу информации на входе. Значение, близкое к 1, означает, что информация важна и должна быть сохранена, в то время как значение, близкое к 0, указывает на то, что информацию можно игнорировать.

<li>Вентиль забывания: Этот вентиль решает, какую информацию следует сохранить из состояния ячейки. Он также использует сигмоидную функцию активации для присвоения значения от 0 до 1 каждому элементу информации в состоянии ячейки. Значение, близкое к 1, указывает на то, что информацию следует сохранить, в то время как значение, близкое к 0, говорит о том, что информацию следует отбросить.

<li>Выходной вентиль: Этот вентиль контролирует, какая часть состояния ячейки должна быть выведена на текущем временном шаге. Он использует сигмоидную функцию активации для присвоения значения от 0 до 1 каждому элементу информации в состоянии ячейки. Значение, близкое к 1, указывает на то, что информацию следует выводить, в то время как значение, близкое к 0, означает, что информацию следует сохранить внутри ячейки.

Эти вентили позволяют LSTM выборочно запоминать или забывать информацию, что делает его эффективным для задач, связанных с последовательностями или временными рядами. Вентили реализованы внутри класса LSTM в PyTorch, поэтому они не определены явно в коде.

В блоках LSTM (Long Short-Term Memory) размер вентилей определяется размером скрытого состояния. Каждый вентиль в блоке LSTM имеет такой же размер, как скрытое состояние. Это означает, что выход входного вентиля, вентиля забывания и выходного вентиля имеют тот же размер, что и скрытое состояние.

```python
import requests


def download_file(url):
    """
    Downloads a file from a URL.

    Returns:
        str: The name of the downloaded file.
    """
    filename = url.split('/')[-1]
    response = requests.get(url)
    with open(filename, 'wb') as f:
        f.write(response.content)
    return filename


url = 'https://raw.githubusercontent.com/samsoberanis/DailyDelhiClimate/main/DailyDelhiClimate.csv'
data = download_file(url=url)
train_df=pd.read_csv(data)
# train_df = train_df.drop(['date'], axis=1)  # Drop the date column
print(f"len(train_df):{len(train_df)}")
train_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Create a MinMaxScaler object
scaler = MinMaxScaler()
# Normalize the data
meantemp = scaler.fit_transform(meantemp.reshape(-1,1))
```

```python
def split_data(data,time_step=12):
    dataX=[]
    datay=[]
    for i in range(len(data)-time_step):
        dataX.append(data[i:i+time_step])
        datay.append(data[i+time_step])
    dataX=np.array(dataX).reshape(len(dataX),time_step,-1)
    datay=np.array(datay)
    return dataX,datay
```

```python
dataX,datay=split_data(meantemp,time_step=12)
print(f"dataX.shape:{dataX.shape},datay.shape:{datay.shape}")
```

```python
# Function to split the dataset into training and testing sets
def train_test_split(dataX, datay, shuffle=True, percentage=0.8):
    """
    Pass the training data X and labels y in the form of numpy.array arrays
    The split ratio is set to training set : test set = 8 : 2
    """
    if shuffle:
        random_num = [index for index in range(len(dataX))]
        np.random.shuffle(random_num)
        dataX = dataX[random_num]
        datay = datay[random_num]
    split_num = int(len(dataX) * percentage)
    train_X = dataX[:split_num]
    train_y = datay[:split_num]
    test_X = dataX[split_num:]
    test_y = datay[split_num:]
    return train_X, train_y, test_X, test_y
```

```python
train_X,train_y,test_X,test_y=train_test_split(dataX,datay,shuffle=False,percentage=0.8)
print(f"train_X.shape:{train_X.shape},test_X.shape:{test_X.shape}")
```

```python
X_train,y_train=train_X,train_y
```

```python
class CNN_LSTM(nn.Module):
    """
    A Convolutional Neural Network (CNN) followed by a Long Short-Term Memory (LSTM) model.

    Attributes:
        conv (nn.Conv1d): A 1D convolutional layer.
        lstm (nn.LSTM): An LSTM layer.
        fc (nn.Linear): A linear layer to produce the final output.

    Args:
        conv_input (int): The number of expected channels in the input to the Conv1d layer.
        input_size (int): The number of expected features in the input to the LSTM layer.
        hidden_size (int): The number of features in the hidden state of the LSTM layer.
        num_layers (int): The number of recurrent layers in the LSTM layer.
        output_size (int): The size of the output from the final linear layer.
    """

    def __init__(self, conv_input, input_size, hidden_size, num_layers, output_size):
        super(CNN_LSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.conv = nn.Conv1d(conv_input, conv_input, 1)
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        """
        Defines the computation performed at every call.

        Args:
            x (torch.Tensor): The input tensor.

        Returns:
            torch.Tensor: The output of the model.
        """
        x = self.conv(x)
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)  # Initialize hidden state h0
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)  # Initialize memory state c0
        out, _ = self.lstm(x, (h0, c0))  # LSTM forward propagation
        out = self.fc(out[:, -1, :])  # Take the output of the last time step as the prediction result
        return out
```

```python
%%time
import matplotlib.pyplot as plt


test_X1=torch.Tensor(test_X)
test_y1=torch.Tensor(test_y)

# Define input, hidden state, and output dimensions
input_size = 1  # Input feature dimension
conv_input=12
hidden_size = 64  # LSTM hidden state dimension
hidden_size_large = hidden_size * 2  # LSTM hidden state dimension
num_layers = 5  # Number of LSTM layers
output_size = 1  # Output dimension (prediction target dimension)

# Create an instance of the CNN_LSTM model
model = CNN_LSTM(conv_input,input_size, hidden_size, num_layers, output_size)
model_large = CNN_LSTM(conv_input, input_size, hidden_size_large, num_layers, output_size)

# Optimizer for the large model
optimizer_large = optim.Adam(model_large.parameters(), lr=0.0001, betas=(0.5, 0.999))

# The training period is 500 times
num_epochs=500
batch_size=64 # Quantity of training at one time
# Optimizer
optimizer=optim.Adam(model.parameters(),lr=0.0001,betas=(0.5,0.999))
# Loss function
criterion=nn.MSELoss()

# Lists to store the losses for the small model
train_losses=[]
test_losses=[]

# Lists to store the losses for the large model
train_losses_large = []
test_losses_large = []

print(f"start")

# Train both models
for epoch in range(num_epochs):
    # Shuffle the training data
    random_num = np.random.permutation(len(train_X))
    train_X = train_X[random_num]
    train_y = train_y[random_num]

    # Convert the training data to tensors
    train_X1 = torch.Tensor(train_X[:batch_size])
    train_y1 = torch.Tensor(train_y[:batch_size])

    # Train the original model
    model.train()
    optimizer.zero_grad()
    output = model(train_X1)
    train_loss = criterion(output, train_y1)
    train_loss.backward()
    optimizer.step()

    # Train the large model
    model_large.train()
    optimizer_large.zero_grad()
    output_large = model_large(train_X1)
    train_loss_large = criterion(output_large, train_y1)
    train_loss_large.backward()
    optimizer_large.step()

    if epoch % 1 == 0:
        # Evaluate the original model
        model.eval()
        with torch.no_grad():
            output = model(test_X1)
            test_loss = criterion(output, test_y1)
        train_losses.append(train_loss.item())
        test_losses.append(test_loss.item())

        # Evaluate the large model
        model_large.eval()
        with torch.no_grad():
            output_large = model_large(test_X1)
            test_loss_large = criterion(output_large, test_y1)
        train_losses_large.append(train_loss_large.item())
        test_losses_large.append(test_loss_large.item())

        if epoch % 50 == 0:
            print(f"epoch: {epoch}, train_loss: {train_loss}, test_loss: {test_loss}, train_loss_large: {train_loss_large}, test_loss_large: {test_loss_large}")

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Training loss 64')
plt.plot(test_losses, label='Testing loss 64')
plt.plot(train_losses_large, label='Training loss 128', linestyle='dashed')
plt.plot(test_losses_large, label='Testing loss 128', linestyle='dashed')
plt.title('Training and Testing Losses Over Time')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Other strategies exist for generating predictions

In the field of time series forecasting, several strategies exist for generating predictions:

<li> Direct Multi-step Forecast: This strategy involves training a distinct model for each future time step. For instance, to predict the subsequent 10 days, 10 different models would be trained - one for each day. Although this approach can be computationally intensive, it occasionally yields more accurate results.

<li> Multiple Output Strategy: This strategy necessitates modifying the model to output multiple future time steps simultaneously. Rather than outputting a single future value, the model could output the next 10 future values. Implementing this strategy requires modifications to the model architecture and can be more complex.

<li> Hybrid Strategy: This strategy represents a fusion of the recursive and direct strategies. Multiple models are trained, each predicting a different number of steps ahead. For example, one model might be trained to predict 1 step ahead, another to predict 2 steps ahead, and so on. This approach can strike a balance between computational efficiency and forecast accuracy.

В области прогнозирования временных рядов существуют несколько стратегий для генерации прогнозов:

<li>Прямой многократный прогноз: Эта стратегия предполагает обучение отдельной модели для каждого будущего временного шага. Например, для прогнозирования следующих 10 дней будут обучены 10 различных моделей - по одной на каждый день. Хотя этот подход может быть вычислительно сложным, иногда он дает более точные результаты.

<li>Стратегия с множественным выводом: Эта стратегия требует модификации модели для одновременного вывода нескольких будущих временных шагов. Вместо вывода одного будущего значения модель может выводить следующие 10 будущих значений. Реализация этой стратегии требует изменений в архитектуре модели и может быть более сложной.

<li>Гибридная стратегия: Эта стратегия представляет собой сочетание рекурсивной и прямой стратегий. Обучаются несколько моделей, каждая из которых прогнозирует разное количество шагов вперед. Например, одна модель может быть обучена для прогнозирования на 1 шаг вперед, другая - на 2 шага вперед и т.д. Этот подход может обеспечить баланс между вычислительной эффективностью и точностью прогноза.
