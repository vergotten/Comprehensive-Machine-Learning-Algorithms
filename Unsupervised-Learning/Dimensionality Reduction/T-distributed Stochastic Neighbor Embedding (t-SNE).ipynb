# T-distributed Stochastic Neighbor Embedding (t-SNE)

t-SNE is a machine learning algorithm for visualization developed by Laurens van der Maaten and Geoffrey Hinton. It is a nonlinear dimensionality reduction technique well-suited for embedding high-dimensional data for visualization in a low-dimensional space of two or three dimensions.

## How it Works

The t-SNE algorithm calculates a probability distribution in the high-dimensional space such that similar objects have a higher probability of being picked while dissimilar points have an extremely small likelihood of being picked. It then calculates a similar probability distribution in the low-dimensional space which minimizes the divergence between the two distributions with respect to the location of the points in the map.

## Use Cases

t-SNE is used in a wide variety of applications, including:

- **Natural Language Processing (NLP)**: t-SNE is used to visualize high-dimensional word embeddings in two or three dimensions, which can reveal semantic relationships between words.

- **Computer Vision**: t-SNE is used to visualize features extracted from images, which can reveal patterns and structures in the image data.

- **Bioinformatics**: t-SNE is used to visualize complex biological data, such as gene expression data, which can reveal patterns of gene expression that are related to different biological conditions.

## Limitations

While t-SNE is a powerful tool for visualization, it does have some limitations:

- **Computational Complexity**: t-SNE has a computational complexity of O(n^2), which makes it less suitable for very large datasets.

- **Hyperparameters**: t-SNE has a few key hyperparameters (such as the perplexity) that can significantly affect the resulting visualizations, and it's not always clear how to choose these parameters.

- **Global Structure**: t-SNE tends to emphasize local structure over global structure, which means that the distances between clusters in the visualization may not have meaningful interpretations.

# T-distributed Stochastic Neighbor Embedding (t-SNE)

t-SNE - это алгоритм машинного обучения для визуализации, разработанный Лоренсом ван дер Маатеном и Джеффри Хинтоном. Это нелинейная техника уменьшения размерности, которая хорошо подходит для встраивания данных высокой размерности для визуализации в пространстве низкой размерности из двух или трех измерений.

## Как это работает

Алгоритм t-SNE вычисляет вероятностное распределение в пространстве высокой размерности таким образом, что похожие объекты имеют большую вероятность быть выбранными, в то время как непохожие точки имеют крайне малую вероятность быть выбранными. Затем он вычисляет аналогичное вероятностное распределение в пространстве низкой размерности, которое минимизирует расхождение между двумя распределениями относительно расположения точек на карте.

## Применение

t-SNE используется в широком спектре приложений, включая:

- **Обработка естественного языка (NLP)**: t-SNE используется для визуализации векторов слов высокой размерности в двух или трех измерениях, что может раскрыть семантические связи между словами.

- **Компьютерное зрение**: t-SNE используется для визуализации признаков, извлеченных из изображений, что может раскрыть шаблоны и структуры в данных изображений.

- **Биоинформатика**: t-SNE используется для визуализации сложных биологических данных, таких как данные о выражении генов, что может раскрыть шаблоны выражения генов, связанные с различными биологическими условиями.

## Ограничения

Несмотря на то, что t-SNE - мощный инструмент для визуализации, у него есть некоторые ограничения:

- **Вычислительная сложность**: t-SNE имеет вычислительную сложность O(n^2), что делает его менее подходящим для очень больших наборов данных.

- **Гиперпараметры**: у t-SNE есть несколько ключевых гиперпараметров (таких как перплексия), которые могут значительно влиять на получаемые визуализации, и не всегда понятно, как выбирать эти параметры.

- **Глобальная структура**: t-SNE склонен подчеркивать локальную структуру за счет глобальной структуры, что означает, что расстояния между кластерами на визуализации могут не иметь значимых интерпретаций.

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

class TSNEVisualization:
    """
    A class used to represent a t-SNE Visualization.

    ...

    Attributes
    ----------
    clf : sklearn.manifold.TSNE
        The t-SNE classifier from scikit-learn.
    digits_data : sklearn.utils.Bunch
        The Digits dataset.
    transformed_data : numpy.ndarray
        The transformed data by the classifier on the Digits data.

    Methods
    -------
    load_data():
        Loads the Digits data.
    transform():
        Transforms the Digits data.
    plot_transformed_data():
        Plots the transformed data.
    """

    def __init__(self, n_components=2, perplexity=30.0):
        """
        Constructs all the necessary attributes for the TSNEVisualization object.
        """

        self.clf = TSNE(n_components=n_components, perplexity=perplexity)

    def load_data(self):
        """Loads the Digits data."""

        self.digits_data = load_digits()

    def transform(self):
        """Transforms the Digits data."""

        self.transformed_data = self.clf.fit_transform(self.digits_data.data)

    def plot_transformed_data(self):
        """Plots the transformed data."""

        plt.scatter(self.transformed_data[:, 0], self.transformed_data[:, 1], c=self.digits_data.target, cmap='viridis')
        plt.colorbar()
        plt.show()

# Example usage:
tsne_visualization = TSNEVisualization()
tsne_visualization.load_data()
tsne_visualization.transform()
tsne_visualization.plot_transformed_data()
```

This code creates a t-SNE visualization for the Digits dataset. The TSNEVisualization class has methods to load the data, transform the data using t-SNE, and plot the resulting transformed data. The classifier is initialized with the parameters n_components and perplexity, which control the dimensionality of the embedded space and the balance between preserving the local and global structure of the data, respectively. The load_data method loads the Digits dataset, the transform method fits the t-SNE transformation to the data and applies it, and the plot_transformed_data method creates a scatter plot of the transformed data points colored by their true label. The example usage at the end of the code demonstrates how to use the class to perform t-SNE visualization on the Digits dataset. The scatter plot shows the data points in the embedded space colored by their true label. The Digits dataset is a good choice for this demonstration because it consists of 8x8 pixel images of digits, and the goal is to visualize the high-dimensional data in two dimensions. The color bar shows the mapping between the colors and the true labels. The scikit-learn documentation provides more information on the options and how to use them. The code is written in Python and uses the scikit-learn library, which provides a wide range of tools for machine learning and data analysis. The matplotlib library is used for creating the scatter plot.

Этот код создает визуализацию t-SNE для набора данных Digits. Класс TSNEVisualization имеет методы для загрузки данных, преобразования данных с использованием t-SNE и построения полученных преобразованных данных. Классификатор инициализируется с параметрами n_components и perplexity, которые контролируют размерность встроенного пространства и баланс между сохранением локальной и глобальной структуры данных, соответственно. Метод load_data загружает набор данных Digits, метод transform применяет преобразование t-SNE к данным и применяет его, а метод plot_transformed_data создает диаграмму рассеяния преобразованных точек данных, окрашенных в соответствии с их истинной меткой. Пример использования в конце кода демонстрирует, как использовать класс для выполнения визуализации t-SNE на наборе данных Digits. Диаграмма рассеяния показывает точки данных во встроенном пространстве, окрашенные в соответствии с их истинной меткой. Набор данных Digits - хороший выбор для этой демонстрации, потому что он состоит из изображений цифр размером 8x8 пикселей, и цель состоит в том, чтобы визуализировать данные высокой размерности в двух измерениях. Цветовая панель показывает соответствие между цветами и истинными метками. Документация scikit-learn предоставляет больше информации о вариантах и о том, как их использовать.
