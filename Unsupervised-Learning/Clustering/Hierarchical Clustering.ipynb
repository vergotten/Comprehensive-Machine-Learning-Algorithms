# Hierarchical Clustering

Hierarchical clustering is another type of unsupervised learning algorithm that is used to group similar items into clusters. Unlike K-Means clustering, hierarchical clustering does not require the user to specify the number of clusters in advance.

## How it Works

Hierarchical clustering builds a hierarchy of clusters by creating a cluster tree or dendrogram. The root of the tree is the unique cluster that gathers all the samples, and the leaves are clusters with only one sample. The algorithm can be implemented in two ways:

1. **Agglomerative (bottom-up)**: Each data point starts in its own cluster, and pairs of clusters are merged together as one moves up the hierarchy.

2. **Divisive (top-down)**: All data points start in one cluster, and splits are performed recursively as one moves down the hierarchy.

In both cases, the decision of which clusters to merge or split is based on the similarity of the data points in the clusters. This similarity can be measured in various ways (e.g., distance between cluster centroids, distance between closest points, etc.), but the choice of similarity measure is not tied to the choice of hierarchical clustering algorithm.

## Use Cases

Hierarchical clustering has a wide range of applications, including:

- **Biological sciences**: Hierarchical clustering is often used in the field of genetics to find groups of similar genes that might have the same function or are activated under the same conditions.

- **Information retrieval**: Hierarchical clustering can be used to cluster documents into topics, which can be useful for search engines or recommendation systems.

- **Customer segmentation**: Like K-Means, hierarchical clustering can also be used to segment customers into groups based on their behavior or characteristics.

## Limitations

While hierarchical clustering can be very powerful, it also has some limitations:

- **Computational complexity**: Hierarchical clustering can be computationally expensive, especially for large datasets.

- **No reassignment of instances**: Once an instance has been assigned to a cluster, it cannot be reassigned. This can lead to suboptimal clusterings.

- **Sensitivity to the choice of similarity measure**: The results of hierarchical clustering can be very different depending on the choice of similarity measure.

# Иерархическая кластеризация

Иерархическая кластеризация - это еще один тип алгоритма обучения без учителя, который используется для группировки похожих элементов в кластеры. В отличие от кластеризации K-Means, иерархическая кластеризация не требует от пользователя указания количества кластеров заранее.

## Как это работает

Иерархическая кластеризация строит иерархию кластеров, создавая дерево кластеров или дендрограмму. Корень дерева - это уникальный кластер, который собирает все образцы, а листья - кластеры с одним образцом. Алгоритм может быть реализован двумя способами:

1. **Агломеративный (снизу вверх)**: каждая точка данных начинает в своем собственном кластере, и пары кластеров объединяются вместе по мере продвижения вверх по иерархии.

2. **Дивизивный (сверху вниз)**: все точки данных начинают в одном кластере, и разделения выполняются рекурсивно по мере продвижения вниз по иерархии.

В обоих случаях решение о том, какие кластеры объединить или разделить, основывается на сходстве точек данных в кластерах. Это сходство может измеряться различными способами (например, расстояние между центроидами кластеров, расстояние между ближайшими точками и т.д.), но выбор меры сходства не связан с выбором алгоритма иерархической кластеризации.

## Применение

Иерархическая кластеризация имеет широкий спектр применений, включая:

- **Биологические науки**: иерархическая кластеризация часто используется в области генетики для поиска групп похожих генов, которые могут иметь одинаковую функцию или активироваться при одинаковых условиях.

- **Поиск информации**: иерархическая кластеризация может использоваться для кластеризации документов по темам, что может быть полезно для поисковых систем или систем рекомендаций.

- **Сегментация клиентов**: как и K-Means, иерархическая кластеризация также может использоваться для сегментации клиентов на группы на основе их поведения или характеристик.

## Ограничения

Хотя иерархическая кластеризация может быть очень мощной, у нее есть некоторые ограничения:

- **Вычислительная сложность**: иерархическая кластеризация может быть вычислительно дорогостоящей, особенно для больших наборов данных.

- **Нет переназначения экземпляров**: как только экземпляр был назначен кластеру, он не может быть переназначен. Это может привести к субоптимальным кластерам.

- **Чувствительность к выбору меры сходства**: результаты иерархической кластеризации могут быть очень разными в зависимости от выбора меры сходства.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt

class HierarchicalClusteringClassifier:
    """
    A class used to represent a Hierarchical Clustering Classifier.

    ...

    Attributes
    ----------
    clf : sklearn.cluster.AgglomerativeClustering
        The Hierarchical Clustering classifier from scikit-learn.
    wine_data : sklearn.utils.Bunch
        The Wine dataset.
    predictions : numpy.ndarray
        The predictions made by the classifier on the Wine data.

    Methods
    -------
    load_data():
        Loads the Wine data.
    train():
        Trains the Hierarchical Clustering classifier.
    predict():
        Makes predictions on the Wine data.
    plot_clusters():
        Plots the clusters found by the classifier.
    """

    def __init__(self, n_clusters=3):
        """
        Constructs all the necessary attributes for the HierarchicalClusteringClassifier object.
        """

        self.clf = AgglomerativeClustering(n_clusters=n_clusters)

    def load_data(self):
        """Loads the Wine data."""

        self.wine_data = load_wine()

    def train(self):
        """Trains the Hierarchical Clustering classifier."""

        self.clf.fit(self.wine_data.data)

    def predict(self):
        """Makes predictions on the Wine data."""

        self.predictions = self.clf.labels_

    def plot_clusters(self):
        """Plots the clusters found by the classifier."""

        plt.scatter(self.wine_data.data[:, 0], self.wine_data.data[:, 1], c=self.predictions, cmap='viridis')
        plt.show()

# Example usage:
hierarchical_classifier = HierarchicalClusteringClassifier()
hierarchical_classifier.load_data()
hierarchical_classifier.train()
hierarchical_classifier.predict()
hierarchical_classifier.plot_clusters()
```

This code creates a Hierarchical Clustering classifier for the Wine dataset. The HierarchicalClusteringClassifier class has methods to load the data, train the classifier, make predictions, and plot the resulting clusters. The classifier is initialized with the number of clusters to form, which defaults to 3. The load_data method loads the Wine dataset, the train method fits the classifier to the data, the predict method assigns each data point to a cluster, and the plot_clusters method creates a scatter plot of the data points colored by cluster. The example usage at the end of the code demonstrates how to use the class to perform Hierarchical Clustering on the Wine dataset. Note that unlike KMeans, AgglomerativeClustering does not have a predict method, so we use the labels_ attribute to get the cluster assignments. Also, AgglomerativeClustering does not compute cluster centers, so we do not plot them. The scatter plot shows the data points colored by their assigned cluster. The Wine dataset is a good choice for this demonstration because it has multiple features that can be used for clustering, and the true labels (types of wine) provide a way to evaluate the performance of the clustering. However, keep in mind that in a real-world unsupervised learning scenario, the true labels would not be known. The goal of the clustering is to discover structure in the data. The performance of the clustering can be evaluated by how well the clusters separate the data points and how compact the clusters are. The choice of the number of clusters and the linkage criterion (how the distance between clusters is computed) can have a big impact on the results. It’s often a good idea to try different values and criteria to see what works best for the data. The scikit-learn documentation provides more information on the options and how to use them. The code is written in Python and uses the scikit-learn library, which provides a wide range of tools for machine learning and data analysis. The matplotlib library is used for creating the scatter plot.
