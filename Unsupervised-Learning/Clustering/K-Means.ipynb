# K-Means Clustering

K-Means is a popular unsupervised learning algorithm used for clustering. The goal of clustering is to partition a dataset into groups, or clusters, of similar items. In the context of K-Means, similarity is defined in terms of distance in the feature space, with items in the same cluster being closer to each other than to items in other clusters.

## How it Works

The K-Means algorithm works by iteratively updating a set of k cluster centers to minimize the total within-cluster variance. The steps are as follows:

1. Initialize k cluster centers randomly.
2. Assign each data point to the nearest cluster center.
3. Update each cluster center by computing the mean of all data points assigned to it.
4. Repeat steps 2 and 3 until convergence (i.e., the cluster assignments no longer change).

## Use Cases

K-Means is used in a wide variety of applications, including:

- **Customer segmentation**: Companies use K-Means to segment their customers into groups based on purchasing behavior, demographics, etc. This can inform targeted marketing strategies.
- **Image compression**: K-Means can be used to quantize color spaces, reducing the number of unique colors in an image and thereby compressing it.
- **Anomaly detection**: Data points that are far from any cluster center can be considered anomalies or outliers. This can be useful in fields like fraud detection or fault detection.

## Limitations

While K-Means is simple and can be very effective, it does have some limitations:

- **Choice of k**: The number of clusters k is a hyperparameter that must be set in advance. There are techniques to choose a good k, but they generally require running the algorithm multiple times.
- **Sensitivity to initialization**: The final solution can be sensitive to the initial random assignment of cluster centers. To mitigate this, K-Means is often run multiple times with different initializations.
- **Assumption of spherical clusters**: K-Means assumes that clusters are spherical and equally sized, which may not always be the case.

# Кластеризация K-Means

K-Means - это популярный алгоритм обучения без учителя, используемый для кластеризации. Цель кластеризации - разделить набор данных на группы или кластеры похожих элементов. В контексте K-Means сходство определяется в терминах расстояния в пространстве признаков, при этом элементы в одном кластере ближе друг к другу, чем к элементам в других кластерах.

## Как это работает

Алгоритм K-Means работает путем итеративного обновления набора центров кластеров k для минимизации общего внутрикластерного разброса. Шаги следующие:

1. Инициализируйте k центров кластеров случайным образом.
2. Назначьте каждую точку данных ближайшему центру кластера.
3. Обновите каждый центр кластера, вычислив среднее значение всех точек данных, назначенных ему.
4. Повторяйте шаги 2 и 3 до сходимости (то есть назначения кластеров больше не меняются).

## Применение

K-Means используется в самых разных приложениях, включая:

- **Сегментация клиентов**: Компании используют K-Means для разделения своих клиентов на группы на основе поведения при покупке, демографии и т.д. Это может информировать стратегии целевого маркетинга.
- **Сжатие изображений**: K-Means может быть использован для квантования цветовых пространств, уменьшая количество уникальных цветов на изображении и тем самым сжимая его.
- **Обнаружение аномалий**: Точки данных, которые находятся далеко от любого центра кластера, могут считаться аномалиями или выбросами. Это может быть полезно в областях, таких как обнаружение мошенничества или обнаружение неисправностей.

## Ограничения

Хотя K-Means прост и может быть очень эффективен, у него есть некоторые ограничения:

- **Выбор k**: Количество кластеров k - это гиперпараметр, который должен быть установлен заранее. Существуют техники выбора хорошего k, но они обычно требуют многократного выполнения алгоритма.
- **Чувствительность к инициализации**: Итоговое решение может быть чувствительным к начальному случайному назначению центров кластеров. Чтобы смягчить это, K-Means часто запускается несколько раз с разными инициализациями.
- **Предположение о сферических кластерах**: K-Means предполагает, что кластеры являются сферическими и имеют одинаковые размеры, что может не всегда быть так.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

class KMeansClassifier:
    """
    A class used to represent a K-Means Classifier.

    ...

    Attributes
    ----------
    clf : sklearn.cluster.KMeans
        The K-Means classifier from scikit-learn.
    iris_data : sklearn.utils.Bunch
        The Iris dataset.
    predictions : numpy.ndarray
        The predictions made by the classifier on the Iris data.

    Methods
    -------
    load_data():
        Loads the Iris data.
    train():
        Trains the K-Means classifier.
    predict():
        Makes predictions on the Iris data.
    plot_clusters():
        Plots the clusters found by the classifier.
    """

    def __init__(self, n_clusters=3):
        """
        Constructs all the necessary attributes for the KMeansClassifier object.
        """

        self.clf = KMeans(n_clusters=n_clusters)

    def load_data(self):
        """Loads the Iris data."""

        self.iris_data = load_iris()

    def train(self):
        """Trains the K-Means classifier."""

        self.clf.fit(self.iris_data.data)

    def predict(self):
        """Makes predictions on the Iris data."""

        self.predictions = self.clf.predict(self.iris_data.data)

    def plot_clusters(self):
        """Plots the clusters found by the classifier."""

        plt.scatter(self.iris_data.data[:, 0], self.iris_data.data[:, 1], c=self.predictions, cmap='viridis')
        centers = self.clf.cluster_centers_
        plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
        plt.show()

# Example usage:
kmeans_classifier = KMeansClassifier()
kmeans_classifier.load_data()
kmeans_classifier.train()
kmeans_classifier.predict()
kmeans_classifier.plot_clusters()
```

The provided code loads the Iris dataset, trains a K-Means classifier on it, and makes predictions for each data point. It then plots the resulting clusters. The number of clusters is set to 3 by default, but this can be modified when creating the KMeansClassifier object. The plot displays the data points colored according to their predicted cluster, with the cluster centers marked in black. Although the Iris dataset has more than two dimensions, only the first two dimensions are plotted for simplicity. For visualizing higher-dimensional data, a dimensionality reduction technique like PCA may be considered.

It's important to note that the K-Means algorithm is sensitive to the initialization of the centroids. This means that different results might be obtained if the code is run multiple times. To achieve consistent results, the random_state parameter can be set when creating the KMeans object. For instance, the line self.clf = KMeans(n_clusters=n_clusters) can be modified to self.clf = KMeans(n_clusters=n_clusters, random_state=42). This ensures that the centroids are initialized in the same way every time the code is run.

Furthermore, K-Means assumes that all clusters are spherical and have similar sizes. This might not be the case for all datasets. If the data does not meet these assumptions, a different clustering algorithm may be considered.

Предоставленный код загружает набор данных Iris, обучает на нем классификатор K-Means и делает прогнозы для каждой точки данных. Затем он строит получившиеся кластеры. Количество кластеров по умолчанию установлено равным 3, но его можно изменить при создании объекта KMeansClassifier. График отображает точки данных, окрашенные в соответствии с их предсказанным кластером, с центрами кластеров, отмеченными черным цветом. Хотя набор данных Iris имеет более двух измерений, для простоты мы строим только первые два измерения. Для визуализации данных большей размерности может быть рассмотрена техника уменьшения размерности, такая как PCA.

Важно отметить, что алгоритм K-Means чувствителен к инициализации центроидов. Это означает, что при многократном запуске кода могут быть получены различные результаты. Для получения согласованных результатов параметр random_state можно установить при создании объекта KMeans. Например, строку self.clf = KMeans(n_clusters=n_clusters) можно изменить на self.clf = KMeans(n_clusters=n_clusters, random_state=42). Это гарантирует, что центроиды инициализируются одинаково каждый раз при запуске кода.

Кроме того, K-Means предполагает, что все кластеры являются сферическими и имеют одинаковые размеры. Это может не всегда быть так во всех наборах данных. Если ваши данные не соответствуют этим предположениям, может быть рассмотрен другой алгоритм кластеризации.
