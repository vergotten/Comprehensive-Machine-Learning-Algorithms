Here is a Python code snippet using PyTorch that generates a continuous uniform distribution:

```python
import torch
import matplotlib.pyplot as plt

# Parameters
a = 0
b = 1

# Create a tensor
tensor = torch.Tensor(1000).uniform_(a, b)

# Plot
plt.hist(tensor.numpy(), bins=30, density=True)
plt.show()
```

Here is a Python code snippet using PyTorch that generates a Normal Distribution:

```python
import torch
import matplotlib.pyplot as plt

# Parameters
mu = 0
sigma = 1

# Create a tensor
tensor = torch.normal(mean=mu, std=sigma, size=(1000,))

# Plot
plt.hist(tensor.numpy(), bins=30, density=True)
plt.show()
```

Here is a Python code snippet using PyTorch that generates a Bernoulli Distribution:

```python
import torch
import matplotlib.pyplot as plt

# Parameter
p = 0.5

# Create a tensor
tensor = torch.bernoulli(torch.full((1000,), p))

# Plot
plt.hist(tensor.numpy(), bins=30, density=True)
plt.show()
```

Here is a Python code snippet using PyTorch that generates a Binomial Distribution:

```python
import torch
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5

# Create a tensor
tensor = torch.distributions.Binomial(n, p).sample((1000,))

# Plot
plt.hist(tensor.numpy(), bins=30, density=True)
plt.show()
```
