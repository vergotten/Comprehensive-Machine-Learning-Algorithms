Here is a Python code snippet using PyTorch that generates a continuous uniform distribution:

```python
import torch
import matplotlib.pyplot as plt

a = 0
b = 1

num_samples = 1000

# uniform_ distribution
normal_d = torch.Tensor(num_samples,).uniform_(a, b); print([x for x in normal_d[:10]])

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_figheight(4)
fig.set_figwidth(15)

ax1.plot(normal_d)
ax1.set_xlabel("Sample Index")
ax1.set_ylabel("Value")

ax2.hist(normal_d, bins=30)
ax2.set_xlabel("Value")
ax2.set_ylabel("Frequency")

plt.show()
```

Here is a Python code snippet using PyTorch that generates a normal distribution:

```python
import torch
import matplotlib.pyplot as plt

mu = 0
sigma = 1

num_samples = 10000

# Normal distribution
normal_d = torch.normal(mean=mu, std=sigma, size=(num_samples,)); print([x for x in normal_d[:10]])

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_figheight(4)
fig.set_figwidth(15)

ax1.plot(normal_d)
ax1.set_xlabel("Sample Index")
ax1.set_ylabel("Value")

ax2.hist(normal_d, bins=30)
ax2.set_xlabel("Value")
ax2.set_ylabel("Frequency")

plt.show()
```

Here is a Python code snippet using PyTorch that generates a bernoulli distribution:

```python
import torch
import matplotlib.pyplot as plt

p = torch.Tensor([0.5])  # 50%

num_samples = 100

# Bernoulli distribution
bernoulli_d = torch.distributions.Bernoulli(probs=p).sample((num_samples,)); print([x for x in bernoulli_d][:10])

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_figheight(4)
fig.set_figwidth(15)

ax1.plot(bernoulli_d[:num_samples])
ax1.set_xlabel("Sample Index")
ax1.set_ylabel("Value")

ax2.hist(bernoulli_d[:num_samples], bins=5)
ax2.set_xlabel("Value")
ax2.set_ylabel("Frequency")

plt.show()
```

Here is a Python code snippet using PyTorch that generates a binomial distribution:

```python
import torch
import matplotlib.pyplot as plt

p = torch.Tensor([0.5])  # 50%

num_samples = 100

# Binominal distribution
binomial_d = torch.distributions.Binomial(num_samples, p).sample((num_samples,)); print([x for x in binomial_d][:10])

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_figheight(4)
fig.set_figwidth(15)

ax1.plot(binomial_d[:num_samples])
ax1.set_xlabel("Sample Index")
ax1.set_ylabel("Value")

ax2.hist(binomial_d[:num_samples], bins=15)
ax2.set_xlabel("Value")
ax2.set_ylabel("Frequency")

plt.show()
```
