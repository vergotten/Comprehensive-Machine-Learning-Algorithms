# Logistic Regression

Logistic Regression is a statistical model that estimates the probability of an event occurring based on a given dataset of independent variables. It is often used for classification and predictive analytics. The outcome is a probability, so the dependent variable is bounded between 0 and 1. In logistic regression, a logit transformation is applied on the odds—that is, the probability of success divided by the probability of failure. This is also commonly known as the log odds, or the natural logarithm of odds.

## Uses of Logistic Regression
Logistic regression is commonly used for prediction and classification problems. Some of these use cases include:
1. **Medical Research**: For example, researchers might want to know how exercise and weight impact the probability of having a heart attack.
2. **Admissions**: Researchers might want to know how GPA, ACT score, and number of AP classes taken impact the probability of getting accepted into a particular university.
3. **Spam Detection**: A business might want to know whether word count and country of origin impact the probability that an email is spam.
4. **Fraud Detection**: A credit card company might want to know whether transaction amount and credit score impact the probability of a given transaction being fraudulent.

## Examples of Logistic Regression
Here are some examples of logistic regression in real life:
1. **NBA Draft**: Suppose a sports data scientist wants to use the predictor variables (1) points, (2) rebounds, and (3) assists to predict the probability that a given college basketball player gets drafted into the NBA.
2. **Spam Detection**: Suppose a business wants to use the predictor variables (1) word count and (2) country of origin to predict the probability that a given email is spam.
3. **Political Preference**: Suppose a political scientist wants to use the predictor variables (1) annual income and (2) years of education to predict the probability that an individual will vote for one of four different presidential candidates.
4. **Sports Preference**: Suppose a sports analyst wants to use the predictor variables (1) TV hours viewed per week and (2) age to predict the probability that an individual will pick either basketball, football, or baseball as their preferred sport.

As a data scientist, understanding logistic regression is crucial as it provides a strong foundation for understanding more complex machine learning algorithms. It's also a powerful tool in its own right for making predictions based on a set of related data points.

# Логистическая регрессия

Логистическая регрессия - это статистическая модель, которая оценивает вероятность наступления события на основе заданного набора независимых переменных. Она часто используется для классификации и прогнозного анализа. Результатом является вероятность, поэтому зависимая переменная ограничена между 0 и 1. В логистической регрессии применяется логит-преобразование шансов, то есть вероятность успеха, деленная на вероятность неудачи. Это также известно как логарифм шансов или натуральный логарифм шансов.

## Применение логистической регрессии
Логистическая регрессия обычно используется для прогнозирования и задач классификации. Некоторые из этих применений включают:
1. **Медицинские исследования**: Например, исследователи могут захотеть узнать, как физическая активность и вес влияют на вероятность сердечного приступа.
2. **Поступление в университет**: Исследователи могут захотеть узнать, как средний балл, результаты теста ACT и количество пройденных классов AP влияют на вероятность поступления в определенный университет.
3. **Обнаружение спама**: Компания может захотеть узнать, влияют ли количество слов и страна происхождения на вероятность того, что электронное письмо является спамом.
4. **Обнаружение мошенничества**: Кредитная компания может захотеть узнать, влияют ли сумма транзакции и кредитный рейтинг на вероятность того, что данная транзакция является мошеннической.

## Примеры логистической регрессии
Вот несколько примеров логистической регрессии в реальной жизни:
1. **НБА Драфт**: Предположим, спортивный аналитик данных хочет использовать предикторные переменные (1) очки, (2) подборы и (3) передачи, чтобы предсказать вероятность того, что данный игрок в колледж-баскетбол будет выбран на драфте НБА.
2. **Обнаружение спама**: Предположим, компания хочет использовать предикторные переменные (1) количество слов и (2) страна происхождения, чтобы предсказать вероятность того, что данное электронное письмо является спамом.
3. **Политические предпочтения**: Предположим, политолог хочет использовать предикторные переменные (1) годовой доход и (2) количество лет образования, чтобы предсказать вероятность того, что человек проголосует за одного из четырех разных кандидатов в президенты.
4. **Спортивные предпочтения**: Предположим, спортивный аналитик хочет использовать предикторные переменные (1) количество часов просмотра телевизора в неделю и (2) возраст, чтобы предсказать вероятность того, что человек выберет баскетбол, футбол или бейсбол в качестве предпочитаемого вида спорта.

Для специалиста по данным понимание логистической регрессии имеет решающее значение, поскольку оно обеспечивает прочную основу для понимания более сложных алгоритмов машинного обучения. Это также мощный инструмент сам по себе для прогнозирования на основе набора связанных точек данных.

```python
import torch
from torch import nn, optim
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np

class LogisticRegressionModel:
    def __init__(self, n_samples=5000, n_features=2, n_classes=2):
        """
        Initialize the LogisticRegressionModel with a synthetic dataset.

        Parameters:
        n_samples (int): The number of samples.
        n_features (int): The number of features.
        n_classes (int): The number of classes.
        """
        X_numpy, y_numpy = make_classification(
            n_samples=n_samples,
            n_features=n_features,
            n_informative=n_features,
            n_redundant=0,
            n_repeated=0,
            n_classes=n_classes,
            random_state=42
        )
        self.X = torch.tensor(X_numpy, dtype=torch.float32)
        self.y = torch.tensor(y_numpy, dtype=torch.float32).view(-1, 1)
        self.model = nn.Linear(self.X.shape[1], self.y.shape[1])
        self.criterion = nn.BCEWithLogitsLoss()
        self.optimizer = optim.SGD(self.model.parameters(), lr=0.01)

    def train(self, epochs=1000):
        """
        Train the model for a number of epochs.

        Parameters:
        epochs (int): The number of epochs to train the model.
        """
        for epoch in range(epochs):
            y_pred = self.model(self.X)
            loss = self.criterion(y_pred, self.y)
            if epoch % 100 == 0:
                print(f'Epoch {epoch}, Loss: {loss.item()}')
            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()

    def plot(self, x_range=None, y_range=None):
        """
        Plot the dataset and the learned decision boundary.

        Parameters:
        x_range (tuple): The lower and upper limits of the x-axis.
        y_range (tuple): The lower and upper limits of the y-axis.
        """
        plt.figure(figsize=(8, 6))
        plt.scatter(self.X.numpy()[:, 0], self.X.numpy()[:, 1], c=self.y.numpy().squeeze(), s=15)
        x_vals = np.array([self.X.min(), self.X.max()])
        y_vals = -(self.model.bias.data.item() + self.model.weight.data.numpy()[0][0]*x_vals) / self.model.weight.data.numpy()[0][1]
        plt.plot(x_vals, y_vals, color='red')
        if x_range is not None:
            plt.xlim(x_range)
        if y_range is not None:
            plt.ylim(y_range)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Synthetic Classification Dataset with Decision Boundary')
        plt.grid(True)
        plt.show()

# Usage:
lr = LogisticRegressionModel()
lr.train()
lr.plot(x_range=(-2, 2), y_range=(-2, 2))
```
