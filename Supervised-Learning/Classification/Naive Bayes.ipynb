# Naive Bayes

Naive Bayes is a simple but surprisingly powerful probabilistic machine learning algorithm used for predictive modeling and classification tasks. It is a collection of classification algorithms based on Bayes’ Theorem. It is not a single algorithm but a family of algorithms where all of them share a common principle, i.e., every pair of features being classified is independent of each other.

## Usage
Naive Bayes is used due to its simplicity, efficiency, and effectiveness in certain types of classification tasks. It’s particularly suitable for text classification, spam filtering, and sentiment analysis. It can handle large volumes of data, including millions of data records. The core of the classifier depends on the Bayes theorem with an assumption of independence among predictors.

## Assumptions
The fundamental Naive Bayes assumption is that each feature makes an independent equal contribution to the outcome. For example, the temperature being ‘Hot’ has nothing to do with the humidity or the outlook being ‘Rainy’ has no effect on the winds. Hence, the features are assumed to be independent. Secondly, each feature is given the same weight (or importance). None of the attributes is irrelevant and assumed to be contributing equally to the outcome.

## Examples
1. **Spam Filtering**: Naive Bayes is commonly used in spam filtering. By analyzing the frequency of certain words in spam emails, the algorithm can predict whether a new email is spam or not.
2. **Sentiment Analysis**: Naive Bayes can be used to determine whether a given piece of text is positive, negative, or neutral. This is often used in social media monitoring, allowing businesses to gain insights into how customers view their products.
3. **Document Classification**: Naive Bayes can be used to classify documents into different categories based on the content of the document. For example, it can be used to automatically tag customer queries into categories like 'billing', 'technical support', 'general inquiry', etc.

Despite its simplicity and the naive assumption of feature independence, the Naive Bayes algorithm often performs well and is widely used because it's easy to understand, can be trained quickly, and is able to handle both continuous and discrete data.

# Наивный Байес

Наивный Байес - это простой, но удивительно мощный вероятностный алгоритм машинного обучения, используемый для предсказательного моделирования и задач классификации. Это совокупность алгоритмов классификации, основанных на теореме Байеса. Это не один алгоритм, а семейство алгоритмов, где все они разделяют общий принцип, то есть каждая пара классифицируемых признаков независима друг от друга.

## Использование
Наивный Байес используется из-за его простоты и эффективности в определенных типах задач классификации. Он особенно подходит для текстовой классификации, фильтрации спама и анализа тональности. Он может обрабатывать большие объемы данных, включая миллионы записей данных. Ядро классификатора зависит от теоремы Байеса с предположением о независимости среди предикторов.

## Предположения
Основное предположение наивного Байеса заключается в том, что каждый признак вносит независимый равный вклад в результат. Например, температура «горячая» никак не связана с влажностью, а прогноз «дождливый» не влияет на ветер. Следовательно, предполагается, что признаки независимы. Во-вторых, каждому признаку присваивается одинаковый вес (или значение). Предполагается, что ни один из атрибутов не является несущественным и в равной степени влияет на результат.

## Примеры
1. **Фильтрация спама**: Наивный Байес широко используется для фильтрации спама. Анализируя частоту определенных слов в спам-письмах, алгоритм может предсказать, является ли новое письмо спамом или нет.
2. **Анализ тональности**: Наивный Байес может использоваться для определения того, является ли данный текст позитивным, негативным или нейтральным. Это часто используется при мониторинге социальных сетей, позволяя компаниям получать информацию о том, как клиенты относятся к их продуктам.
3. **Классификация документов**: Наивный Байес может использоваться для классификации документов в разные категории на основе содержания документа. Например, он может использоваться для автоматической маркировки запросов клиентов по категориям, таким как «биллинг», «техническая поддержка», «общий запрос» и т.д.

Несмотря на свою простоту и наивное предположение о независимости признаков, алгоритм наивного Байеса часто работает хорошо и широко используется, потому что его легко понять, можно быстро обучить и он способен обрабатывать как непрерывные, так и дискретные данные.

```python
import torch
from torchvision import datasets, transforms
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class NBClassifier:
    """
    A class used to represent a Naive Bayes (NB) Classifier.

    ...

    Attributes
    ----------
    clf : sklearn.naive_bayes.GaussianNB
        The NB classifier from scikit-learn.
    X_train : numpy.ndarray
        The training data.
    y_train : numpy.ndarray
        The training labels.
    X_test : numpy.ndarray
        The test data.
    y_test : numpy.ndarray
        The test labels.
    predictions : numpy.ndarray
        The predictions made by the classifier on the test data.

    Methods
    -------
    load_data():
        Loads the MNIST data.
    train():
        Trains the NB classifier.
    predict():
        Makes predictions on the test data.
    accuracy():
        Returns the accuracy of the classifier.
    plot_confusion_matrix():
        Plots a confusion matrix of the classifier's predictions.
    """

    def __init__(self):
        """
        Constructs all the necessary attributes for the NBClassifier object.
        """

        self.clf = GaussianNB()

    def load_data(self):
        """Loads the MNIST data."""

        # Define a transform to normalize the data
        transform = transforms.Compose([transforms.ToTensor(),
                                        transforms.Normalize((0.5,), (0.5,))])

        # Download and load the training data
        trainset = datasets.MNIST('~/.pytorch/MNIST_data/', download=True, train=True, transform=transform)
        testset = datasets.MNIST('~/.pytorch/MNIST_data/', download=True, train=False, transform=transform)

        # Use DataLoader to handle batching of the data
        trainloader = torch.utils.data.DataLoader(trainset, batch_size=len(trainset), shuffle=True)
        testloader = torch.utils.data.DataLoader(testset, batch_size=len(testset), shuffle=True)

        # Convert the data to numpy arrays
        traindata = next(iter(trainloader))
        testdata = next(iter(testloader))

        self.X_train = traindata[0].numpy().reshape(traindata[0].shape[0], -1)
        self.y_train = traindata[1].numpy()

        self.X_test = testdata[0].numpy().reshape(testdata[0].shape[0], -1)
        self.y_test = testdata[1].numpy()

    def train(self):
        """Trains the NB classifier."""

        self.clf.fit(self.X_train, self.y_train)

    def predict(self):
        """Makes predictions on the test data."""

        self.predictions = self.clf.predict(self.X_test)

    def accuracy(self):
        """Returns the accuracy of the classifier."""

        return accuracy_score(self.y_test, self.predictions)

    def plot_confusion_matrix(self):
        """Plots a confusion matrix of the classifier's predictions."""

        cm = confusion_matrix(self.y_test, self.predictions)
        plt.figure(figsize=(10,10))
        sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
        plt.ylabel('Actual label');
        plt.xlabel('Predicted label');
        plt.title('Confusion Matrix', size = 15);
        plt.show()

# Example usage:
nb_classifier = NBClassifier()
nb_classifier.load_data()
nb_classifier.train()
nb_classifier.predict()
print("Accuracy:", nb_classifier.accuracy())
nb_classifier.plot_confusion_matrix()
```

Naive Bayes classifiers assume that the value of a particular feature is independent of the value of any other feature, given the class variable. This assumption is called class conditional independence. It is made to simplify the computation and in this sense considered to be “Naive”.

Классификаторы наивного Байеса предполагают, что значение определенного признака не зависит от значения любого другого признака, учитывая классовую переменную. Это предположение называется условной независимостью класса. Оно делается для упрощения вычислений и в этом смысле считается “наивным”.

Naive Bayes classifiers perform well on datasets where the number of features is large compared to the number of samples. They are particularly effective for text classification, document classification, and spam filtering.

```python
import torch
from torchvision import datasets, transforms
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class NBClassifier:
    """
    A class used to represent a Naive Bayes (NB) Classifier.

    ...

    Attributes
    ----------
    clf : sklearn.naive_bayes.GaussianNB
        The NB classifier from scikit-learn.
    X_train : numpy.ndarray
        The training data.
    y_train : numpy.ndarray
        The training labels.
    X_test : numpy.ndarray
        The test data.
    y_test : numpy.ndarray
        The test labels.
    predictions : numpy.ndarray
        The predictions made by the classifier on the test data.

    Methods
    -------
    load_data():
        Loads the FashionMNIST data.
    train():
        Trains the NB classifier.
    predict():
        Makes predictions on the test data.
    accuracy():
        Returns the accuracy of the classifier.
    plot_confusion_matrix():
        Plots a confusion matrix of the classifier's predictions.
    """

    def __init__(self):
        """
        Constructs all the necessary attributes for the NBClassifier object.
        """

        self.clf = GaussianNB()

    def load_data(self):
        """Loads the FashionMNIST data."""

        # Define a transform to normalize the data
        transform = transforms.Compose([transforms.ToTensor(),
                                        transforms.Normalize((0.5,), (0.5,))])

        # Download and load the training data
        trainset = datasets.FashionMNIST('~/.pytorch/FashionMNIST_data/', download=True, train=True, transform=transform)
        testset = datasets.FashionMNIST('~/.pytorch/FashionMNIST_data/', download=True, train=False, transform=transform)

        # Use DataLoader to handle batching of the data
        trainloader = torch.utils.data.DataLoader(trainset, batch_size=len(trainset), shuffle=True)
        testloader = torch.utils.data.DataLoader(testset, batch_size=len(testset), shuffle=True)

        # Convert the data to numpy arrays
        traindata = next(iter(trainloader))
        testdata = next(iter(testloader))

        self.X_train = traindata[0].numpy().reshape(traindata[0].shape[0], -1)
        self.y_train = traindata[1].numpy()

        self.X_test = testdata[0].numpy().reshape(testdata[0].shape[0], -1)
        self.y_test = testdata[1].numpy()

    def train(self):
        """Trains the NB classifier."""

        self.clf.fit(self.X_train, self.y_train)

    def predict(self):
        """Makes predictions on the test data."""

        self.predictions = self.clf.predict(self.X_test)

    def accuracy(self):
        """Returns the accuracy of the classifier."""

        return accuracy_score(self.y_test, self.predictions)

    def plot_confusion_matrix(self):
        """Plots a confusion matrix of the classifier's predictions."""

        cm = confusion_matrix(self.y_test, self.predictions)
        plt.figure(figsize=(10,10))
        sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
        plt.ylabel('Actual label');
        plt.xlabel('Predicted label');
        plt.title('Confusion Matrix', size = 15);
        plt.show()

# Example usage:
nb_classifier = NBClassifier()
nb_classifier.load_data()
nb_classifier.train()
nb_classifier.predict()
print("Accuracy:", nb_classifier.accuracy())
nb_classifier.plot_confusion_matrix()
```
