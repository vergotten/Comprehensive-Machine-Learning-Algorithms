# CatBoost

CatBoost, short for Categorical Boosting, is an open-source boosting library developed by Yandex. It is designed for use on problems like regression and classification having a very large number of independent features. CatBoost is a variant of gradient boosting that can handle both categorical and numerical features.

## Key Features

- **Handling Categorical Features**: CatBoost can handle categorical features without any feature encoding.
- **Handling Missing Values**: Unlike other Models, CatBoost can easily handle any missing values in the dataset.
- **Automatic Feature Scaling**: CatBoost internally scales all the columns to the same scaling.
- **Built-in Cross-validation**: CatBoost internally applies a cross-validation method to choose the best hyperparameters for the model.
- **Regularizations**: CatBoost supports both L1 and L2 regularization methods to reduce overfitting.
- **Language Support**: It can be used in both Python and R language.

## Unique Techniques

CatBoost uses an algorithm called symmetric weighted quantile sketch (SWQS) which automatically handles the missing values in the dataset to reduce overfitting and improve the overall performance of the dataset.

## Use Cases

CatBoost can be used for both regression and classification problems. Here are a few examples:

- **Predictive Maintenance**: CatBoost can be used to predict the probability of a device failure at a future point in time based on various predictors such as temperature, voltage, rotation speed, etc.

- **Fraud Detection**: CatBoost can be used to identify fraudulent transactions in real-time based on historical transaction patterns.

- **Customer Churn Prediction**: CatBoost can help predict which customers are likely to churn in the future based on their past behavior and interactions.

- **Medical Diagnosis**: CatBoost can be used to predict disease outcomes based on various patient characteristics and test results.

Она предназначена для использования в задачах регрессии и классификации, имеющих очень большое количество независимых признаков. CatBoost - это вариант градиентного усиления, который может обрабатывать как категориальные, так и числовые признаки.

## Основные особенности

- **Обработка категориальных признаков**: CatBoost может обрабатывать категориальные признаки без кодирования признаков.
- **Обработка пропущенных значений**: В отличие от других моделей, CatBoost может легко обрабатывать любые пропущенные значения в наборе данных.
- **Автоматическое масштабирование признаков**: CatBoost внутренне масштабирует все столбцы до одного и того же масштаба.
- **Встроенная кросс-валидация**: CatBoost внутренне применяет метод кросс-валидации для выбора лучших гиперпараметров для модели.
- **Регуляризации**: CatBoost поддерживает методы регуляризации L1 и L2 для снижения переобучения.
- **Поддержка языков**: Он может быть использован как на Python, так и на R.

## Уникальные техники

CatBoost использует алгоритм, называемый симметричный взвешенный квантильный эскиз (SWQS), который автоматически обрабатывает пропущенные значения в наборе данных для снижения переобучения и улучшения общей производительности набора данных.

## Примеры использования

CatBoost можно использовать как для задач регрессии, так и для задач классификации. Вот несколько примеров:

- **Прогнозное обслуживание**: CatBoost можно использовать для прогнозирования вероятности отказа устройства в будущем времени на основе различных предикторов, таких как температура, напряжение, скорость вращения и т. д.

- **Обнаружение мошенничества**: CatBoost можно использовать для идентификации мошеннических транзакций в реальном времени на основе исторических шаблонов транзакций.

- **Прогнозирование оттока клиентов**: CatBoost может помочь предсказать, какие клиенты, вероятно, уйдут в будущем, на основе их прошлого поведения и взаимодействия.

- **Медицинская диагностика**: CatBoost можно использовать для прогнозирования исходов болезней на основе различных характеристик пациента и результатов тестов.

```python
try:
    from catboost import CatBoostClassifier
    from sklearn.model_selection import train_test_split
    from sklearn.datasets import load_wine
    from sklearn.metrics import accuracy_score, confusion_matrix
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
except ImportError as e:
    !pip install catboost
    print(f"Error: {e}. Please install the required libraries.")

class CatBoost_Classifier:
    def __init__(self):
        self.clf = CatBoostClassifier(verbose=0, n_estimators=100)

    def load_data(self):
        wine = load_wine()
        self.X = wine.data
        self.y = wine.target
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=0.2, random_state=42)

    def train(self):
        self.clf.fit(self.X_train, self.y_train)

    def predict(self):
        self.predictions = self.clf.predict(self.X_test)

    def accuracy(self):
        return accuracy_score(self.y_test, self.predictions)

    def plot_confusion_matrix(self):
        cm = confusion_matrix(self.y_test, self.predictions)
        plt.figure(figsize=(10,10))
        sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
        plt.ylabel('Actual label');
        plt.xlabel('Predicted label');
        plt.title('Confusion Matrix', size = 15);
        plt.show()

catboost_classifier = CatBoost_Classifier()
catboost_classifier.load_data()
catboost_classifier.train()
catboost_classifier.predict()
print("Accuracy:", catboost_classifier.accuracy())
catboost_classifier.plot_confusion_matrix()
```
