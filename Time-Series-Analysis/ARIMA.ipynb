# ARIMA: Autoregressive Integrated Moving Average

ARIMA stands for Autoregressive Integrated Moving Average. It is a statistical model used in time series analysis. The model is a generalization of the Autoregressive Moving Average (ARMA) model.

## Components of ARIMA

- 'Autoregressive' (AR) part indicates that the variable of interest is regressed on its own lagged, or prior, values.
- 'Integrated' (I) part indicates that the data values have been replaced with the difference between their values and the previous values.
- 'Moving Average' (MA) part indicates that the regression error is actually a linear combination of error terms whose values occurred contemporaneously and at various times in the past.

Non-seasonal ARIMA models are generally denoted ARIMA(p, d, q) where parameters p, d, and q are non-negative integers, p is the order (number of time lags) of the autoregressive model, d is the degree of differencing (the number of times the data have had past values subtracted), and q is the order of the moving-average model.

ARIMA models can be estimated following the Box–Jenkins approach. They are applied in cases where data show evidence of non-stationarity, where an initial differencing step can be applied one or more times to eliminate the non-stationarity.

## Use Cases of ARIMA

ARIMA models are typically used to model time series data. Time series data is data that consists of many repeated measurements that are taken on the same quantity. These measurements are typically taken over and over again over a long period of time.

Some of the real-world use cases of ARIMA include:

- Forecasting the quantity of a good needed for the next time period based on historical data.
- Forecasting sales and interpreting seasonal changes in sales.
- Estimating the impact of marketing events, new product launches, and so on.
- A hospital in Singapore accurately predicted the number of beds they will be needing in 3 days during the SARS epidemic using ARIMA.
- Modeling and forecasting demand in a food company using ARIMA.

## Examples of ARIMA

ARIMA model is a class of linear models that utilizes historical values to forecast future values. For example, an ARIMA model might seek to predict a stock's future prices based on its past performance or forecast a company's earnings based on past periods.

ARIMA makes use of lagged moving averages to smooth time series data. They are widely used in technical analysis to forecast future security prices. Autoregressive models implicitly assume that the future will resemble the past.

## Conclusion

ARIMA is a powerful tool for time series forecasting. It is widely used in various fields for forecasting based on historical data. However, it requires a good understanding of the underlying assumptions and proper tuning of the parameters for optimal performance. It is also important to note that ARIMA assumes that the future will resemble the past, which may not always be the case in real-world scenarios. Therefore, it is always recommended to use ARIMA in conjunction with other forecasting methods for more robust and accurate predictions.

# ARIMA: Авторегрессионная интегрированная скользящая средняя

ARIMA - это модель авторегрессионной интегрированной скользящей средней. Это статистическая модель, используемая в анализе временных рядов. Модель является обобщением модели авторегрессионной скользящей средней (ARMA).

## Компоненты ARIMA

- Часть 'Авторегрессионная' (AR) указывает на то, что интересующая нас переменная регрессируется по своим собственным запаздывающим или предыдущим значениям.
- Часть 'Интегрированная' (I) указывает на то, что значения данных заменены разностью между их значениями и предыдущими значениями.
- Часть 'Скользящая средняя' (MA) указывает на то, что ошибка регрессии на самом деле является линейной комбинацией ошибок, значения которых возникли одновременно и в разное время в прошлом.

Несезонные модели ARIMA обычно обозначаются как ARIMA(p, d, q), где параметры p, d и q являются неотрицательными целыми числами, p - порядок (количество временных задержек) авторегрессионной модели, d - степень дифференцирования (количество раз, когда из данных вычитались прошлые значения), и q - порядок модели скользящей средней.

Модели ARIMA можно оценивать, следуя подходу Бокса-Дженкинса. Они применяются в случаях, когда данные показывают признаки нестационарности, где начальный шаг дифференцирования может быть применен один или несколько раз для устранения нестационарности.

## Применение ARIMA

Модели ARIMA обычно используются для моделирования данных временных рядов. Данные временных рядов - это данные, которые состоят из многих повторяющихся измерений, взятых на одной и той же величине. Эти измерения обычно проводятся снова и снова в течение длительного периода времени.

Некоторые из реальных примеров использования ARIMA включают:

- Прогнозирование количества товара, необходимого на следующий период времени, на основе исторических данных.
- Прогнозирование продаж и интерпретация сезонных изменений в продажах.
- Оценка влияния маркетинговых мероприятий, запуска новых продуктов и т.д.
- Больница в Сингапуре точно предсказала количество кроватей, которые им понадобятся через 3 дня во время эпидемии SARS, используя ARIMA.
- Моделирование и прогнозирование спроса в пищевой компании с помощью ARIMA.

## Примеры ARIMA

Модель ARIMA - это класс линейных моделей, которые используют исторические значения для прогнозирования будущих значений. Например, модель ARIMA может стремиться предсказать будущие цены на акции на основе их прошлой производительности или прогнозировать прибыль компании на основе прошлых периодов.

ARIMA использует запаздывающие скользящие средние для сглаживания данных временных рядов. Они широко используются в техническом анализе для прогнозирования будущих цен на ценные бумаги. Авторегрессионные модели неявно предполагают, что будущее будет похоже на прошлое.

## Заключение

ARIMA - это мощный инструмент для прогнозирования временных рядов. Он широко используется в различных областях для прогнозирования на основе исторических данных. Однако он требует хорошего понимания основных предположений и правильной настройки параметров для оптимальной производительности. Также важно отметить, что ARIMA предполагает, что будущее будет похоже на прошлое, что может не всегда быть так в реальных сценариях. Поэтому всегда рекомендуется использовать ARIMA в сочетании с другими методами прогнозирования для более надежных и точных прогнозов.

```python
try:
    from statsmodels.tsa.arima.model import ARIMA
    from sklearn.datasets import load_diabetes
    import matplotlib.pyplot as plt
except ImportError:
    !pip install statsmodels scikit-learn matplotlib

class ARIMAModel:
    """
    A class used to represent an ARIMA Model.

    ...

    Attributes
    ----------
    model : statsmodels.tsa.arima.model.ARIMA
        The ARIMA model from statsmodels.
    diabetes_data : sklearn.utils.Bunch
        The Diabetes dataset.
    fitted_model : statsmodels.tsa.arima.model.ARIMAResultsWrapper
        The fitted model on the Diabetes data.

    Methods
    -------
    load_data():
        Loads the Diabetes data.
    fit():
        Fits the ARIMA model on the Diabetes data.
    plot_predictions():
        Plots the predictions of the fitted model.
    """

    def __init__(self, order=(1, 1, 1)):
        """
        Constructs all the necessary attributes for the ARIMAModel object.
        """

        self.order = order

    def load_data(self):
        """Loads the Diabetes data."""

        self.diabetes_data = load_diabetes()

    def fit(self):
        """Fits the ARIMA model on the Diabetes data."""

        self.model = ARIMA(self.diabetes_data.target, order=self.order)
        self.fitted_model = self.model.fit()

    def plot_predictions(self):
        """Plots the predictions of the fitted model."""

        plt.plot(self.diabetes_data.target)
        plt.plot(self.fitted_model.predict(), color='red')
        plt.show()

# Example usage:
arima_model = ARIMAModel()
arima_model.load_data()
arima_model.fit()
arima_model.plot_predictions()
```

The diabetes dataset utilized in the provided code is not a time series dataset. It serves as an illustrative example. For actual use, it is recommended to replace this with a genuine time series dataset.

The order of the ARIMA model (p, d, q) should be selected based on the specific characteristics of the dataset. The provided code assumes an ARIMA(1, 1, 1) model, which may not be the optimal choice for all datasets. Techniques such as Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots can be employed to determine the appropriate order for the model.

Lastly, it is crucial to verify the residuals of the fitted model to ensure they do not violate any of the model’s assumptions. If violations are detected, consideration should be given to using a different model or transforming the data.

Используемый в предоставленном коде набор данных о диабете не является набором данных временных рядов. Он служит иллюстративным примером. Для фактического использования рекомендуется заменить его на подлинный набор данных временных рядов.

Порядок модели ARIMA (p, d, q) следует выбирать на основе конкретных характеристик набора данных. Предоставленный код предполагает модель ARIMA(1, 1, 1), которая может не быть оптимальным выбором для всех наборов данных. Можно использовать такие методы, как функция автокорреляции (ACF) и частичная функция автокорреляции (PACF), чтобы определить подходящий порядок для модели.
